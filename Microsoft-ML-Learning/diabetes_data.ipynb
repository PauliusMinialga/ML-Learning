## Diabetes data

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn import datasets, linear_model, model_selection
```

```python
#we got all data and it has 442 items shapped in arrays of 10 elements
X, y = datasets.load_diabetes(return_X_y=True)
print(X.shape)
print(X[0])
```

```python
#from the cell above we grab 1 element from each item
X = X[:, 4]
X = X.reshape((-1,1))
print(X.shape)
print(X[0])
```

```python
#splitting data into testing and training sets
X_train, X_test, y_train, y_test = model_selection.train_test_split(X, y, test_size=0.33)
```

```python
#training our linear regression model with the X and y training sets
model = linear_model.LinearRegression()
model.fit(X_train, y_train)
```

```python
#creating a prediction using test data, with the help of predict() function
y_pred = model.predict(X_test)
```

```python
#showing the data in a plot
plt.scatter(X_test, y_test, color='black')
plt.plot(X_test, y_pred, color='blue', linewidth=3)
plt.xlabel('Scaled BMIs')
plt.ylabel('Disease Progression')
plt.title('A Graph Plot Showing Diabetes Progression Against BMI')
plt.show()
```
